## Create a streamlit app example


```
%%writefile app.py
import streamlit as st
import requests

# Page config
st.set_page_config(page_title="MoodMate", layout="centered")

# Styling
st.markdown("""
    <style>
    .stApp {
        background-color: #fff0f5;
        font-family: 'Comic Sans MS', cursive, sans-serif;
        color: black !important;
    }
    .title {
        font-size: 40px;
        font-weight: bold;
        text-align: center;
        color: #ff69b4;
    }
    .footer {
        text-align: center;
        color: #ff69b4;
        font-size: 16px;
        margin-top: 2em;
    }
    .stButton>button {
        color: white !important;
        background-color: #ff69b4 !important;
        border: none;
        font-weight: bold;
    }
    </style>
""", unsafe_allow_html=True)

# Title
st.markdown("<div class='title'>ðŸ’– MoodMate â€“ What's the Vibe Today?</div>", unsafe_allow_html=True)
st.subheader("âœ¨ Drop your feels, bestie âœ¨")

# Emoji mood picker
mood_options = {
    "ðŸ˜© Major Ick": 1,
    "ðŸ˜• Meh": 3,
    "ðŸ™‚ Chill": 5,
    "ðŸ˜Š Cute vibes": 7,
    "ðŸ’ƒ Main character energy": 9,
    "ðŸ¤© Slay all day": 10
}
mood_label = st.selectbox("Pick your vibe today:", list(mood_options.keys()))
mood = mood_options[mood_label]

# --- Movie Recommender ---
def recommend_movie(mood):
    if mood <= 3:
        return "Clueless"
    elif mood <= 5:
        return "To All the Boys Iâ€™ve Loved Before"
    elif mood <= 7:
        return "XO, Kitty"
    elif mood <= 8:
        return "Barbie (2023)"
    else:
        return "Mean Girls (2024)"

if st.button("ðŸ’« Hit me with a rec"):
    title = recommend_movie(mood)
    st.success(f"OMG you def need to watch: **{title}**")

# --- Affirmation API ---
def fetch_affirmation():
    try:
        response = requests.get("https://zenquotes.io/api/random")
        if response.status_code == 200:
            quote = response.json()[0]["q"]
            author = response.json()[0]["a"]
            return f"â€œ{quote}â€ â€” *{author}*"
        else:
            return "Keep glowing. Even clouds gotta move someday."
    except:
        return "You are enough. Always have been. Period."

# Daily Affirmation
st.markdown("---")
st.subheader("ðŸ’¬ Today's Vibe Check Affirmation")
st.markdown("âœ¨ _Get your soul snack. One quote at a time._")
st.info(fetch_affirmation())

# Journaling
st.markdown("---")
st.subheader("ðŸ“ Wanna Journal?")
journal_input = st.text_area("Spill the tea, journal style:", height=200)

if journal_input:
    st.success("Period. Writing = healing.")
    st.download_button(
        label="ðŸ“¥ Save My Vibe Check",
        data=journal_input,
        file_name="my_moodmate_entry.txt",
        mime="text/plain",
    )

# Lofi
st.markdown("---")
st.subheader("ðŸŒ™ Lo-fi for the Soul")
st.video("https://www.youtube.com/watch?v=jfKfPfyJRdk")

# Insta Vibes
st.markdown("ðŸ“¸ _Screenshot your vibe and tag us @MoodMateApp to get featured!_")

# Footer
st.markdown("<div class='footer'>Made with love & âœ¨ main character energy âœ¨ by your fave facilitator ðŸ’•</div>", unsafe_allow_html=True)
```
